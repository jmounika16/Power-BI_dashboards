```python
#checking if there are any outliers(extreme/inconsistent values) in the data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def detect_outliers1(data):
    outliers = []
    threshold = 3

    # Convert data to a numpy array for efficient computation
    data = np.array(data)

    # Filter out non-numeric data (strings, etc.)
    numeric_data = data[np.isreal(data)]

    mean = np.mean(numeric_data)
    std = np.std(numeric_data)

    for i in numeric_data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers

# Example usage
# outlier_pt = detect_outliers(crash_data['CRASH_MONTH'])
```

```python

from scipy.stats import kruskal  # Non-parametric statistical test


# Filter out 'OTHER' and 'UNKNOWN' road surface conditions
filtered_crash_data = crash_data[~crash_data['ROADWAY_SURFACE_COND'].isin(['OTHER', 'UNKNOWN'])]

# Group by 'ROADWAY_SURFACE_COND' and calculate the average severity score
surface_severity = filtered_crash_data.groupby('ROADWAY_SURFACE_COND')['SEVERITY_SCORE'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='ROADWAY_SURFACE_COND', y='SEVERITY_SCORE', data=surface_severity, palette='coolwarm')
plt.title('Average Severity of Crashes by Road Surface Condition')
plt.xlabel('Road Surface Condition')
plt.ylabel('Average Severity Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# To statistically test for a difference in crash severity across different road surface conditions,
# we'll use the Kruskal-Wallis H-test which is the non-parametric version of ANOVA.

# We need to create a list of crash severities for each road surface condition
severity_scores_by_surface = [group['SEVERITY_SCORE'].values for name, group in filtered_crash_data.groupby('ROADWAY_SURFACE_COND')]

# Perform the Kruskal-Wallis H-test
stat, p_value = kruskal(*severity_scores_by_surface)
print(f"Kruskal-Wallis H-test statistic: {stat}, p-value: {p_value}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is a statistically significant difference in crash severity across different road surface conditions.")
else:
    print("There is no statistically significant difference in crash severity across different road surface conditions.")
```
